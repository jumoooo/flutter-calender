CalendarWidget schedule-only UX tweak: hide weekday header, keep drag handle visible.

- Requirements:
  - In scheduleOnly mode, weekday row should not be visible.
  - Drag handle (gray bar) must remain visible at the top of the schedule sheet even when fully expanded.
  - Month label switches to `M.d.요일` only in scheduleOnly, reverts when leaving.
- Implementation notes:
  - Hide weekday row via `_viewMode` check instead of overlapping.
  - Keep schedule `top` clamped to `>= 0.0` so handle never leaves viewport.
  - Use `_viewMode` conditional inside `_buildScheduleColumn` to hide date/lunar line in scheduleOnly.
